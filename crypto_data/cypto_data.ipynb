```python

# Binance 거래소 객체 생성
exchange = ccxt.binance()

# 비트코인/USDT의 과거 데이터를 일별로 가져옵니다.
symbol = 'BTC/USDT'
timeframe = '1d'  # 1일 데이터
since = exchange.parse8601('2020-01-01T00:00:00Z')  # 2020년 1월 1일부터 데이터를 가져옵니다.

ohlcv = exchange.fetch_ohlcv(symbol, timeframe, since)

# 데이터를 판다스 데이터프레임으로 변환합니다.
df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

print(df)
```

```python
import requests

def get_usd_tickers():
    url = "https://api.binance.com/api/v3/ticker/price"
    response = requests.get(url)
    tickers = response.json()
    
    usd_tickers = [ticker for ticker in tickers if ticker['symbol'].endswith('USD')]
    
    return usd_tickers

# Get and print USD tickers
usd_tickers = get_usd_tickers()
for ticker in usd_tickers:
    print(f"Symbol: {ticker['symbol']}, Price: {ticker['price']}")
```

```python
import requests
import json
from datetime import datetime, timedelta

# API 엔드포인트 및 API 키 설정
api_key = 'YOUR_API_KEY'
url_histoday = 'https://min-api.cryptocompare.com/data/histoday'
url_symbols = 'https://min-api.cryptocompare.com/data/all/coinlist'

# 시작일과 종료일 설정
start_date = datetime(2017, 12, 1)
end_date = datetime(2023, 8, 28)

# 모든 암호화폐 티커 리스트 가져오기
response = requests.get(url_symbols, params={'api_key': api_key})
if response.status_code == 200:
    coin_list = response.json()['Data']
    crypto_symbols = list(coin_list.keys())
else:
    print(f"Error fetching coin list: {response.status_code}")
    crypto_symbols = []

# 데이터 저장 딕셔너리 초기화
all_crypto_data = {}

for symbol in crypto_symbols:
    print(f"Fetching data for {symbol}...")
    all_data = []
    current_date = end_date

    while current_date >= start_date:
        params = {
            'fsym': symbol,
            'tsym': 'KRW',
            'limit': 2000,  # 최대 2000개의 데이터 포인트
            'api_key': api_key,
            'toTs': int(current_date.timestamp())
        }

        # API 요청 보내기
        response = requests.get(url_histoday, params=params)
        
        if response.status_code == 200:
            data = response.json()
            if 'Data' in data and len(data['Data']) > 0:
                all_data.extend(data['Data'])
                
                # 가장 오래된 데이터 포인트의 날짜 가져오기
                first_data_point = data['Data'][0]['time']
                current_date = datetime.fromtimestamp(first_data_point) - timedelta(days=1)
            else:
                print(f"No data for {symbol} in KRW")
                break
        else:
            print(f"Error: {response.status_code} for {symbol}")
            break

    if all_data:
        # 중복 데이터 제거 및 정렬
        all_data = {d['time']: d for d in all_data}
        all_data = list(all_data.values())
        all_data.sort(key=lambda x: x['time'])
        
        # 데이터 저장
        all_crypto_data[symbol] = all_data

# 데이터 출력 (선택 사항)
for symbol, data in all_crypto_data.items():
    print(f"Data for {symbol}:")
    print(json.dumps(data, indent=4))

# 데이터 저장 (선택 사항)
with open('all_crypto_krw_data.json', 'w') as f:
    json.dump(all_crypto_data, f, indent=4)
```

```python
import requests
import pandas as pd
from datetime import datetime

def fetch_monthly_volume(symbol, vs_currency, start_date, end_date):
    url = f"https://api.coingecko.com/api/v3/coins/{symbol}/market_chart/range"
    params = {
        'vs_currency': vs_currency,
        'from': start_date.timestamp(),
        'to': end_date.timestamp()
    }
    response = requests.get(url, params=params)
    data = response.json()
    if 'total_volumes' not in data:
        raise KeyError(f"'total_volumes' not found in response for {symbol} in {vs_currency}")
    return data['total_volumes']

def convert_to_monthly_data(volumes):
    df = pd.DataFrame(volumes, columns=['timestamp', 'volume'])
    df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('date', inplace=True)
    monthly_data = df.resample('M').sum()
    return monthly_data

start_date = datetime(2017, 1, 1)
end_date = datetime(2023, 12, 31)

symbols = ['bitcoin']
currencies = ['krw', 'usd']

all_data = {}

for symbol in symbols:
    for currency in currencies:
        try:
            volumes = fetch_monthly_volume(symbol, currency, start_date, end_date)
            monthly_data = convert_to_monthly_data(volumes)
            all_data[f'{symbol}_{currency}'] = monthly_data
        except KeyError as e:
            print(e)

result = pd.concat(all_data.values(), keys=all_data.keys(), axis=1)
result.to_csv('cryptocurrency_monthly_volumes.csv')
```
