### Bithumb

```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# 빗썸 API URL
ohlcv_url = "https://api.bithumb.com/public/candlestick/{currency}_KRW/24h"

# 암호화폐 리스트 가져오기
currency_url = "https://api.bithumb.com/public/ticker/ALL_KRW"
response = requests.get(currency_url)
currencies = response.json()['data'].keys()

# OHLCV 데이터 수집 함수
def get_ohlcv(currency):
    api_url = ohlcv_url.format(currency=currency)
    response = requests.get(api_url)
    data = response.json()['data']
    ohlcv_data = []
    for entry in data:
        timestamp, open_, close, high, low, volume = entry
        date = datetime.fromtimestamp(timestamp / 1000)
        ohlcv_data.append({
            'currency': currency,
            'date': date,
            'open': float(open_),
            'high': float(high),
            'low': float(low),
            'close': float(close),
            'volume': float(volume)
        })
    return ohlcv_data

# 데이터 프레임 생성
start_date = datetime(2017, 12, 1)
end_date = datetime(2023, 8, 28)
all_ohlcv_data = []

for currency in currencies:
    if currency != "date":
        print(f"Fetching OHLCV data for {currency}...")
        ohlcv_data = get_ohlcv(currency)
        # 기간 필터링
        filtered_data = [d for d in ohlcv_data if start_date <= d['date'] <= end_date]
        all_ohlcv_data.extend(filtered_data)

df = pd.DataFrame(all_ohlcv_data)

# 데이터 저장
df.to_csv('bithumb_ohlcv_data.csv', index=False)
print("Data has been saved to bithumb_ohlcv_data.csv")
```

### CoinOne

```python
import requests
import pandas as pd
from datetime import datetime, timezone
import time

# 코인원 티커 정보 조회 API URL
tickers_url = "https://api.coinone.co.kr/public/v2/ticker_utc_new/KRW"

# 코인원 캔들 차트 조회 API URL (기본 틀)
ohlcv_url_template = "https://api.coinone.co.kr/public/v2/chart/KRW/{target_currency}"

# OHLCV 데이터 수집 함수
def get_ohlcv(target_currency, start_date, end_date, interval="1d"):
    url = ohlcv_url_template.format(target_currency=target_currency)
    params = {
        'interval': interval,
        'size': 500  # 최대 500개의 캔들 수집
    }
    all_ohlcv_data = []
    last_timestamp = None
    while True:
        if last_timestamp:
            params['timestamp'] = last_timestamp
        try:
            response = requests.get(url, params=params, timeout=10)
            response.raise_for_status()
            data = response.json().get('chart', [])
            if not data:
                break
            for entry in data:
                timestamp = entry['timestamp']
                date = datetime.fromtimestamp(int(timestamp) / 1000, tz=timezone.utc)  # Unix time을 datetime으로 변환
                if date < start_date:
                    return all_ohlcv_data
                if date > end_date:
                    continue
                all_ohlcv_data.append({
                    'currency': target_currency,
                    'date': date,
                    'open': float(entry['open']),
                    'high': float(entry['high']),
                    'low': float(entry['low']),
                    'close': float(entry['close']),
                    'volume': float(entry['target_volume']),
                    'quote_volume': float(entry['quote_volume'])
                })
            last_timestamp = data[-1]['timestamp'] - 1  # 마지막 타임스탬프를 업데이트
            if data[-1]['timestamp'] < int(start_date.timestamp()) * 1000:
                break
        except (requests.exceptions.HTTPError, requests.exceptions.RequestException, ValueError) as e:
            print(f"Error fetching data for {target_currency}: {e}")
            break
        time.sleep(1)  # 요청 간에 1초 대기
    return all_ohlcv_data

# 티커 정보를 조회하여 KRW로 거래되는 암호화폐 리스트를 가져오는 함수
def get_krw_tickers():
    try:
        response = requests.get(tickers_url, timeout=10)
        response.raise_for_status()
        tickers_data = response.json().get('tickers', [])
        krw_tickers = [ticker['target_currency'] for ticker in tickers_data]
        return krw_tickers
    except (requests.exceptions.HTTPError, requests.exceptions.RequestException, ValueError) as e:
        print(f"Error fetching tickers: {e}")
        return []

# 데이터 수집 및 저장
start_date = datetime(2017, 12, 1, tzinfo=timezone.utc)
end_date = datetime(2023, 8, 28, tzinfo=timezone.utc)
all_ohlcv_data = []

# KRW로 거래되는 암호화폐 리스트를 가져옴
currencies = get_krw_tickers()

# 각 암호화폐의 OHLCV 데이터를 수집
for currency in currencies:
    print(f"Fetching OHLCV data for {currency}...")
    ohlcv_data = get_ohlcv(currency, start_date, end_date)
    if ohlcv_data:
        all_ohlcv_data.extend(ohlcv_data)

# 데이터프레임 생성 및 저장
if all_ohlcv_data:
    df = pd.DataFrame(all_ohlcv_data)
    df.to_csv('coinone_ohlcv_data.csv', index=False)
    print("Data has been saved to coinone_ohlcv_data.csv")
else:
    print("No valid OHLCV data found.")
```

```python
import requests

# 코빗의 티커 정보 조회 API URL
tickers_url = "https://api.korbit.co.kr/v1/ticker/detailed/all"

# 코빗에서 거래되는 모든 KRW 마켓의 티커를 가져오는 함수
def get_krw_tickers():
    try:
        response = requests.get(tickers_url)
        response.raise_for_status()
        tickers_data = response.json()
        krw_tickers = [ticker for ticker in tickers_data.keys() if ticker.endswith('_krw')]
        return krw_tickers
    except Exception as e:
        print(f"Error fetching tickers: {e}")
        return []

# KRW로 거래되는 모든 암호화폐의 티커 리스트를 가져옴
krw_tickers = get_krw_tickers()

# 티커 리스트 출력
print(krw_tickers)
```

```python
import pykorbit
import requests
import pandas as pd
from datetime import datetime

# 데이터 수집 범위 설정
start_date = "2017-12-01"
end_date = "2023-08-28"

# OHLCV 데이터를 수집하는 함수
def get_ohlcv_data(ticker, start, end):
    try:
        base_ticker = ticker.replace('_krw', '')
        data = pykorbit.get_ohlc(base_ticker, start=start, end=end)
        if data is not None and not data.empty:
            data['currency'] = base_ticker
            return data
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
    return pd.DataFrame()


# 모든 티커의 OHLCV 데이터를 수집하여 병합
all_ohlcv_data = pd.DataFrame()
for ticker in krw_tickers:
    print(f"Fetching OHLCV data for {ticker}...")
    ohlcv_data = get_ohlcv_data(ticker, start_date, end_date)
    if not ohlcv_data.empty:
        all_ohlcv_data = pd.concat([all_ohlcv_data, ohlcv_data])

# 데이터프레임 생성 및 저장
if not all_ohlcv_data.empty:
    all_ohlcv_data.to_csv('korbit_ohlcv_data.csv', index=False)
    print("Data has been saved to korbit_ohlcv_data.csv")
else:
    print("No valid OHLCV data found.")
```

```python
import json
from datetime import datetime
import pandas as pd

# Load the JSON data from the file
file_path = 'krw_markets_data.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Define the date range for filtering
start_date = datetime.strptime("2017-12-01", "%Y-%m-%d")
end_date = datetime.strptime("2023-08-28", "%Y-%m-%d")

# Define a function to process all cryptocurrencies based on the keys in the data dictionary
def process_all_cryptos(data, start_date, end_date):
    all_cryptos_data = {}
    for crypto_key in data.keys():
        filtered_data = {}
        for entry in data[crypto_key]:
            date_time = datetime.strptime(entry["time_human_readable"], "%Y-%m-%d %H:%M:%S")
            if start_date <= date_time <= end_date:
                date_str = date_time.strftime("%Y-%m-%d")
                filtered_data[date_str] = filtered_data.get(date_str, 0) + entry["volumeto"]
        all_cryptos_data[crypto_key] = filtered_data
    return all_cryptos_data

# Process data for all cryptocurrencies in the file
all_crypto_data = process_all_cryptos(data, start_date, end_date)

# Combine all data into a single DataFrame
dates = sorted(set(date for crypto_data in all_crypto_data.values() for date in crypto_data.keys()))
combined_data = {
    "Date": dates
}
combined_data.update({
    crypto: [crypto_data.get(date, 0) for date in dates] for crypto, crypto_data in all_crypto_data.items()
})

df_all_cryptos = pd.DataFrame(combined_data)
df_all_cryptos.set_index('Date', inplace=True)
df_all_cryptos

```
