```python
import requests
from bs4 import BeautifulSoup
import re
from tqdm import tqdm

def get_contents(yahoo):
    contents = []

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

    for url in tqdm(yahoo['url']):
        try:
                response = requests.get(url, headers=headers)
                soup = BeautifulSoup(response.text, "html.parser")
                content = soup.select("article > div > div > div > div > div > div.caas-content-wrapper > div.caas-body > p")
                cleaned_content = re.sub('<[^>]*>', '', ''.join(str(item) for item in content))
                contents.append(cleaned_content)
        except:
                contents.append('')

    return contents
```

```python
# Adding the content to the DataFrame
yahoo['content_full'] = get_contents(yahoo)
yahoo
```
