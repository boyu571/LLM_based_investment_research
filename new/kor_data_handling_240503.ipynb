```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'G:\내 드라이브\DATASET\Kakao\kor_sentiment\score_full.csv', engine='python')
df2 = pd.read_csv(r'G:\내 드라이브\DATASET\Kakao\kor_sentiment\score_title.csv', engine='python')
df3 = pd.read_csv(r'G:\내 드라이브\DATASET\Kakao\kor_sentiment\score_content.csv', engine='python')
```

```python
print(len(df1))
print(len(df2))
print(len(df3))
```

```python
filtered_df1 = df1[df1['score2'].notna() & (df1['score2'] >= -1) & (df1['score2'] <= 1)]
filtered_df2 = df2[df2['score_title2'].notna() & (df2['score_title2'] >= -1) & (df2['score_title2'] <= 1)]
filtered_df3 = df3[df3['score_content2'].notna() & (df3['score_content2'] >= -1) & (df3['score_content2'] <= 1)]
```

```python
print(len(filtered_df1))
print(len(filtered_df2))
print(len(filtered_df3))
```

```python
excluded_presses = [
    "CNBC", "Reuters", "Investopedia", "Kiplinger's Personal Finance",
    "CBC News", "Nakedcapitalism.com", "Itsecuritynews.info", "Medium"
]

# 'press' 열이 excluded_presses 리스트에 없는 행만 필터링
filtered_df1 = filtered_df1[~filtered_df1['press'].isin(excluded_presses)]
filtered_df2 = filtered_df2[~filtered_df2['press'].isin(excluded_presses)]
filtered_df3 = filtered_df3[~filtered_df3['press'].isin(excluded_presses)]

print(len(filtered_df1))
print(len(filtered_df2))
print(len(filtered_df3))
```
